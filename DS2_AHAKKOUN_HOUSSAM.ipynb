```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Données PIB et croissance 2024
data = {
    "Pays": ["États-Unis", "Chine", "Inde", "Japon", "Allemagne", "Royaume-Uni", "France", "Canada", "Brésil", "Russie"],
    "Région": ["Amérique du Nord", "Asie", "Asie", "Asie", "Europe", "Europe", "Europe", "Amérique du Nord", "Amérique du Sud", "Europe-Asie"],
    "PIB_2024": [26185, 21643, 3509, 5015, 4200, 3450, 3000, 2100, 1850, 1750],
    "Croissance_2024": [2.1, 4.7, 6.5, 1.2, 0.3, 1.1, 1.0, 0.9, 0.5, 0.8]
}

df = pd.DataFrame(data)

# Synthèse régionale
region_summary = df.groupby("Région").agg({
    "PIB_2024": "sum",
    "Croissance_2024": "mean"
}).reset_index()

sns.set_theme(style="whitegrid")

# 1) Bar chart - PIB par pays
plt.figure(figsize=(12,6))
sns.barplot(x="Pays", y="PIB_2024", data=df, palette="Blues_d")
plt.title("PIB estimé par pays en 2024 (en milliards USD)")
plt.xticks(rotation=45)
plt.ylabel("PIB (milliards $)")
plt.xlabel("Pays")
plt.tight_layout()
plt.show()

# 2) Bar chart - Croissance par pays
plt.figure(figsize=(12,6))
sns.barplot(x="Pays", y="Croissance_2024", data=df, palette="Reds_d")
plt.title("Taux de croissance économique estimé par pays en 2024 (%)")
plt.xticks(rotation=45)
plt.ylabel("Croissance (%)")
plt.xlabel("Pays")
plt.tight_layout()
plt.show()

# 3) Combinaison barres PIB total et ligne croissance moyenne par région
fig, ax1 = plt.subplots(figsize=(10,6))

sns.barplot(x="Région", y="PIB_2024", data=region_summary, ax=ax1, color='skyblue')
ax1.set_ylabel('PIB total par région (milliards $)', color='blue')
ax1.set_xlabel('Région')
ax1.tick_params(axis='y', labelcolor='blue')

ax2 = ax1.twinx()
sns.pointplot(x="Région", y="Croissance_2024", data=region_summary, ax=ax2, color='red')
ax2.set_ylabel('Croissance moyenne (%)', color='red')
ax2.tick_params(axis='y', labelcolor='red')

plt.title("PIB total et croissance moyenne par région en 2024")
plt.tight_layout()
plt.show()

# 4) Scatter plot : Corrélation entre PIB et croissance, par région
plt.figure(figsize=(8,6))
sns.scatterplot(x="PIB_2024", y="Croissance_2024", hue="Région", data=df, s=100, palette="dark")
plt.title("Corrélation entre PIB et taux de croissance (%) en 2024 par pays")
plt.xlabel("PIB 2024 (milliards $)")
plt.ylabel("Croissance 2024 (%)")
plt.legend(title='Région')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Données PIB 2024 estimées et croissance (%)
data = {
    "Pays": ["États-Unis", "Chine", "Inde", "Japon", "Allemagne",
             "Royaume-Uni", "France", "Canada", "Brésil", "Russie"],
    "Région": ["Amérique du Nord", "Asie", "Asie", "Asie", "Europe",
               "Europe", "Europe", "Amérique du Nord", "Amérique du Sud", "Europe-Asie"],
    "PIB_2024": [26185, 21643, 3509, 5015, 4200, 3450, 3000, 2100, 1850, 1750],
    "Croissance_2024": [2.1, 4.7, 6.5, 1.2, 0.3, 1.1, 1.0, 0.9, 0.5, 0.8]
}

df = pd.DataFrame(data)

# Prévision PIB 2025 simple (PIB 2024 * (1 + croissance_2024/100))
df["PIB_2025_Prevu"] = df["PIB_2024"] * (1 + df["Croissance_2024"] / 100)

# Synthèse régionale 2024 et 2025
region_summary_2024 = df.groupby("Région").agg({
    "PIB_2024": "sum",
    "Croissance_2024": "mean"
}).reset_index()

region_summary_2025 = df.groupby("Région").agg({
    "PIB_2025_Prevu": "sum"
}).reset_index()

# Affichage synthèse textuelle
print("Analyse PIB et Croissance 2024 par pays :\n")
for i, row in df.iterrows():
    print(f"- {row['Pays']} ({row['Région']}): PIB 2024 = {row['PIB_2024']:.2f} milliards $, "
          f"Croissance = {row['Croissance_2024']:.2f} %, Prévision PIB 2025 = {row['PIB_2025_Prevu']:.2f} milliards $")

print("\nSynthèse régionale :")
region_report = region_summary_2024.merge(region_summary_2025, on="Région")
for _, row in region_report.iterrows():
    print(f"- {row['Région']}: PIB total 2024 = {row['PIB_2024']:.2f} milliards $, "
          f"Croissance moyenne = {row['Croissance_2024']:.2f} %, PIB total prévu 2025 = {row['PIB_2025_Prevu']:.2f} milliards $")

# Visualisations analytiques
sns.set_theme(style="whitegrid")

# 1) PIB par pays
plt.figure(figsize=(12,6))
sns.barplot(x="Pays", y="PIB_2024", data=df, palette="Blues_d")
plt.title("PIB estimé par pays en 2024 (en milliards $)")
plt.xticks(rotation=45)
plt.ylabel("PIB (milliards $)")
plt.xlabel("Pays")
plt.tight_layout()
plt.show()

# 2) Croissance économique par pays
plt.figure(figsize=(12,6))
sns.barplot(x="Pays", y="Croissance_2024", data=df, palette="Reds_d")
plt.title("Taux de croissance économique estimé par pays en 2024 (%)")
plt.xticks(rotation=45)
plt.ylabel("Croissance (%)")
plt.xlabel("Pays")
plt.tight_layout()
plt.show()

# 3) PIB total et croissance moyenne par région (barres + ligne)
fig, ax1 = plt.subplots(figsize=(10,6))
sns.barplot(x="Région", y="PIB_2024", data=region_summary_2024, ax=ax1, color='skyblue')
ax1.set_ylabel('PIB total 2024 (milliards $)', color='blue')
ax1.set_xlabel('Région')
ax1.tick_params(axis='y', labelcolor='blue')

ax2 = ax1.twinx()
sns.pointplot(x="Région", y="Croissance_2024", data=region_summary_2024, ax=ax2, color='red')
ax2.set_ylabel('Croissance moyenne 2024 (%)', color='red')
ax2.tick_params(axis='y', labelcolor='red')

plt.title("PIB total et croissance moyenne par région en 2024")
plt.tight_layout()
plt.show()

# 4) Corrélation PIB - Croissance par pays
plt.figure(figsize=(8,6))
sns.scatterplot(x="PIB_2024", y="Croissance_2024", hue="Région", data=df, s=100, palette="dark")
plt.title("Corrélation entre PIB 2024 et croissance (%) par pays")
plt.xlabel("PIB 2024 (milliards $)")
plt.ylabel("Croissance 2024 (%)")
plt.legend(title='Région')
plt.tight_layout()
plt.show()
```
